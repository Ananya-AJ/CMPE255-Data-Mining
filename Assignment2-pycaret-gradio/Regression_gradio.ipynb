```python
train_data = pd.read_csv('/content/train.csv')
train_data.shape
train_data.dtypes
#train_data.isnull()
#train_data.fillna(0)
test_data = pd.read_csv('/content/test.csv')
train_data.replace([np.inf, -np.inf], np.nan,inplace=True)
test_data.replace([np.inf, -np.inf], np.nan,inplace=True)



```

Data preprocessing using pycaret's setup

```python
data_setup = setup(data = train_data, target = 'SalePrice', session_id=123,
                   ignore_features = ['Alley','PoolQC','MiscFeature','Fence','FireplaceQu','Utilities'],
                   transformation = True, transformation_method = 'yeo-johnson',remove_multicollinearity = True,
                  normalize = True,remove_outliers= True, transform_target = True, 
                  combine_rare_levels = True,ignore_low_variance = True, use_gpu = True,
                  log_experiment = True, experiment_name = 'houseprice') 
```

```python
compare_models()
```

```python
huber = create_model('huber')
bayesian_ridge = create_model('br')
```

Create Model

```python
huber = tune_model(huber)
bayesian_ridge = tune_model(bayesian_ridge)
```

Blending above models

```python
blender = blend_models(estimator_list = [huber, bayesian_ridge])
```

Predict test data

```python
model = finalize_model(blender)
predictions = predict_model(model, data = test_data)
predictions.head()
```

plot the model

```python
plot_model(model)
```

```python
plot_model(model, plot = 'error')
```

```python
evaluate_model(model)
```
