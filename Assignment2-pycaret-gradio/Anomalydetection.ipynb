```python
import pandas as pd
import kaggle
import numpy as np

train_data = pd.read_csv('/content/Participants_Data_WH18/Train.csv')
test_data = pd.read_csv('/content/Participants_Data_WH18/Test.csv')
train_data.shape
test_data.shape


train_data.replace([np.inf, -np.inf], np.nan,inplace=True)
test_data.replace([np.inf, -np.inf], np.nan,inplace=True)
test_data.drop(columns = ['feature_1', 'feature_2', 'feature_3'], inplace = True)
```

```python
from pycaret.anomaly import *

exp_ano101 = setup(train_data, normalize = True, 
                   ignore_features = ['feature_1', 'feature_2', 'feature_3','Class'],
                   session_id = 123)
```

create models

```python
iforest = create_model('iforest')
print(iforest)
```

```python
svm = create_model('svm', fraction = 0.025)
print(svm)
```

```python
models()
```

```python
iforest_results = assign_model(iforest)
iforest_results.head()
```

plot model

Uniform manifold approximation and projection

```python
plot_model(iforest, plot = 'umap')
```

Predict on unseen data

```python
unseen_predictions = predict_model(iforest, data=test_data)
unseen_predictions.head()
```

Anomaly score labeling for training data too

```python

data_predictions = predict_model(iforest, data = train_data)
data_predictions.head()
```
